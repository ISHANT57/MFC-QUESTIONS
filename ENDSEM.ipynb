```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.start = None

    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.start
        self.start = new_node

    def insert_at_end(self, data):
        new_node = Node(data)
        if self.start is None:
            self.start = new_node
            return
        temp = self.start
        while temp.next:
            temp = temp.next
        temp.next = new_node

    def insert_at_position(self, data, pos):
        if pos == 0:
            self.insert_at_beginning(data)
            return
        new_node = Node(data)
        temp = self.start
        for _ in range(pos - 1):
            if temp is None:
                return
            temp = temp.next
        new_node.next = temp.next
        temp.next = new_node

    def delete_from_beginning(self):
        if self.start:
            self.start = self.start.next

    def delete_from_end(self):
        if self.start is None or self.start.next is None:
            self.start = None
            return
        temp = self.start
        while temp.next.next:
            temp = temp.next
        temp.next = None

    def delete_at_position(self, pos):
        if pos == 0:
            self.delete_from_beginning()
            return
        temp = self.start
        for _ in range(pos - 1):
            if temp is None or temp.next is None:
                return
            temp = temp.next
        temp.next = temp.next.next

    def count(self):
        count = 0
        temp = self.start
        while temp:
            count += 1
            temp = temp.next
        return count

    def reverse(self):
        prev = None
        current = self.start
        while current:
            nxt = current.next
            current.next = prev
            prev = current
            current = nxt
        self.start = prev

    def sort(self):
        end = None
        while end != self.start:
            p = self.start
            while p.next != end:
                q = p.next
                if p.data > q.data:
                    p.data, q.data = q.data, p.data
                p = p.next
            end = p

    # Display the list
    def display(self):
        temp = self.start
        while temp:
            print(temp.data, end=" → ")
            temp = temp.next
        print("None")


# Function call example
print("==== SINGLY LINKED LIST ====")
ll = SinglyLinkedList()

# Inserting elements at the end
ll.insert_at_end(15)
ll.insert_at_end(5)
ll.insert_at_end(25)
ll.insert_at_end(10)

print("Original List:")
ll.display()

# Sorting the list
ll.sort()

print("Sorted List:")
ll.display()
```

```python
def reverse_linked_list(head):
    prev = None
    current = head

    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node

    return prev  # New head of the reversed list

# Example Usage
ll = SinglyLinkedList()
ll.insert_at_end(1)
ll.insert_at_end(2)
ll.insert_at_end(3)
ll.start = reverse_linked_list(ll.start)
ll.display()  # Output: 3 -> 2 -> 1 -> None
```

```python
def detect_cycle(head):
    slow, fast = head, head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            return True  # Cycle detected

    return False  # No cycle detected

# Example Usage
ll = LinkedList()
ll.insert_at_end(1)
ll.insert_at_end(2)
ll.insert_at_end(3)

# Creating a cycle for testing
ll.head.next.next.next = ll.head

print("Cycle Detected:", detect_cycle(ll.head))  # Output: Cycle Detected: True
```

```python
class Node:
    def __init__(self, item=None, next=None):
        self.item = item
        self.next = next

class CLL:
    def __init__(self, last=None):
        self.last = last

    def is_empty(self):
        return self.last is None

    def display(self):
        if self.is_empty():
            print("Linked list is empty")
            return
        temp = self.last.next  # head is last.next
        while temp != self.last:
            print(temp.item, end=" -> ")
            temp = temp.next
        print(temp.item)  # printing last node
    
    def insert_at_start(self, data):
        n = Node(data)
        if self.is_empty():
            n.next = n  # Points to itself (circular)
            self.last = n
        else:
            n.next = self.last.next
            self.last.next = n
            
    def insert_at_last(self, data):
        n = Node(data)
        if self.is_empty():
            n.next = n  # Circular reference
            self.last = n
        else:
            n.next = self.last.next  # Correctly linking new node to head
            self.last.next = n  # Last node points to new node
            self.last = n  # Update last node reference

    def search(self, data):
        if self.is_empty():
            return None
        temp = self.last.next
        while temp != self.last:
            if temp.item == data:
                return temp
            temp = temp.next
        if temp.item == data:  # Checking last node
            return temp
        return None
    
    def insert_after(self, temp, data):
        if temp is not None:
            n = Node(data)
            n.next = temp.next
            temp.next = n
            if temp == self.last:
                self.last = n
                
    def insert_after_pos(self, pos, data):
        if self.is_empty():
            print("List is empty")
            return
        temp = self.last.next  # Start at head
        for i in range(pos):
            if temp == self.last:
                print("Position not found")
                return
            temp = temp.next
        n = Node(data)
        n.next = temp.next
        temp.next = n
        if temp == self.last:
            self.last = n
            
    def delete_first(self):
        if self.is_empty():
            print("List is empty")
            return
        if self.last.next == self.last:  # Only one node exists
            self.last = None
        else:
            self.last.next = self.last.next.next
        
    def delete_last(self):
        if self.is_empty():
            print("CLL is empty")
            return
        if self.last.next == self.last:
            self.last = None
        else:
            temp = self.last.next  # Start from head
            while temp.next != self.last:
                temp = temp.next
            temp.next = self.last.next  # Bypass the last node
            self.last = temp  # Update last pointer
        
    def delete_after_pos(self, pos):
        if self.is_empty():
            print("CLL is empty")
            return
        temp = self.last.next  # Start at head
        for i in range(pos):
            if temp == self.last:
                print("Position not found")
                return
            temp = temp.next
        if temp.next == temp:
            print("Only one node present")
            return
        if temp.next == self.last:
            self.last = temp
        temp.next = temp.next.next

    def rotate(self, k):
        """
        Rotates the circular linked list to the left by k positions.
        """
        if self.is_empty() or self.last.next == self.last:
            return

        # First, determine the length of the list.
        current = self.last.next  # starting from head
        count = 1
        while current != self.last:
            current = current.next
            count += 1
        
        # Effective rotations (in case k > count)
        r = k % count
        if r == 0:
            return  # No rotation needed
        
        # Move 'last' pointer r times
        for _ in range(r):
            self.last = self.last.next
    
    def rotate1(self, k):
        """Rotates by k positions, but only if k is odd."""
        if self.is_empty() or self.last.next == self.last:
            return 

        current = self.last.next
        count = 1
        while current != self.last:
            current = current.next
            count += 1

        r = k % count  # Ensure k is within range
        

        for _ in range(r):
            self.last = self.last.next

# -------------------- Testing the Rotate Function --------------------

# Create a circular linked list and insert elements.
cll = CLL()
cll.insert_at_start(1)
cll.insert_at_last(2)
cll.insert_at_last(3)

print("Original Circular Linked List:")
cll.display()  # Expected: 1 -> 2 -> 3

# Rotate the list by 2 positions.
cll.rotate(2)
print("\nCircular Linked List after rotating left by 2 positions:")
cll.display()  # Expected: 3 -> 1 -> 2

# Rotate the list by 3 more positions.
cll.rotate1(1)
print("\nCircular Linked List after rotating left by 1 more positions (odd k):")
cll.display()  # Expected: 1 -> 2 -> 3 (back to original)
```

```python
def next_greater_element(nums):
    result = []

    for i in range(len(nums)):
        found = False
        for j in range(i + 1, len(nums)):
            if nums[j] > nums[i]:
                result.append(nums[j])
                found = True
                break
        if not found:
            result.append(-1)

    return result

# Test
nums = [6, 9,8, 1, 12, 4]
print("Next Greater Element:", next_greater_element(nums))
```

```python
# ✅ Undo/Redo Functionality Using Stack
class UndoRedoStack:
    def __init__(self):
        self.undo_stack = StackArray()
        self.redo_stack = StackArray()

    def perform_action(self, action):
        print(f"Action performed: {action}")
        self.undo_stack.push(action)
        self.redo_stack = StackArray()  # Clear redo stack

    def undo(self):
        if not self.undo_stack.is_empty():
            action = self.undo_stack.pop()
            self.redo_stack.push(action)
            print(f"Undo: {action}")
        else:
            print("Nothing to undo.")

    def redo(self):
        if not self.redo_stack.is_empty():
            action = self.redo_stack.pop()
            self.undo_stack.push(action)
            print(f"Redo: {action}")
        else:
            print("Nothing to redo.")

    def show_stacks(self):
        print("Undo Stack:")
        self.undo_stack.display()
        print("Redo Stack:")
        self.redo_stack.display()

print("\n=== Undo/Redo Simulation ===")
browser = UndoRedoStack()
browser.perform_action("Open Google")
browser.perform_action("Open YouTube")
browser.perform_action("Open ChatGPT")
browser.show_stacks()

browser.undo()
browser.show_stacks()

browser.undo()
browser.show_stacks()

browser.redo()
browser.show_stacks()
```

```python
def remove_duplicates(arr):
    unique = []
    for item in arr:
        is_duplicate = False
        for u in unique:
            if item == u:
                is_duplicate = True
                break
        if not is_duplicate:
            unique.append(item)
    return unique

# Example usage:
arr = [1, 2, 3, 2, 4, 1, 5, 3]
print("Original Array:", arr)

new_arr = remove_duplicates(arr)
print("Array after removing duplicates:", new_arr)
```

```python
def reverse_array(arr, start, end):
    if start >= end:
        return
    arr[start], arr[end] = arr[end], arr[start]
    reverse_array(arr, start+1, end-1)

arr = [1, 2, 3, 4, 5]
reverse_array(arr, 0, len(arr)-1)
print(arr)  # Output: [5, 4, 3, 2, 1]
```

```python
def rev(arr):
    left=0
    right=len(arr)-1
    
    while right>left:
        arr[left],arr[right]=arr[right],arr[left]
        left+=1
        right-=1
    return arr

rev([2,3,1,9,6,4,0])
```

```python
def max_subarray_sum(arr):
    max_sum = current_sum = arr[0]

    for num in arr[1:]:
        current_sum = max(num, current_sum + num)
        max_sum = max(max_sum, current_sum)
        
    return max_sum

# Example Usage
arr = [-2, -3, 4, -1, -2, 1, 5, -3]
print("Maximum Subarray Sum:", max_subarray_sum(arr))
```

```python
class QueueUsingStacks:
    def __init__(self):
        self.stack1 = []
        self.stack2 = []

    def enqueue(self, x):
        self.stack1.append(x)

    def dequeue(self):
        if not self.stack2:
            while self.stack1:
                self.stack2.append(self.stack1.pop())
        return self.stack2.pop()

queue = QueueUsingStacks()
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
print(queue.dequeue())  # Output: 1
print(queue.dequeue())  # Output: 2
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.front = None
        self.rear = None
        
    def is_empty(self):
        return self.front is None
    
    def enqueue(self, data):
        n = Node(data)
        if self.rear is None:
            self.front = self.rear = n
        else:  
            self.rear.next = n
            self.rear = n
    
    def dequeue(self):
        if self.is_empty():
            return "Queue is empty"
        
        temp = self.front.data
        self.front = self.front.next
        
        if self.front is None:
            self.rear = None
            
        return temp
            
    def display(self):
        if self.is_empty():
            print("Queue is empty")  # Print instead of return
            return
        
        curr = self.front
        while curr:
            print(curr.data, end=" -> ")
            curr = curr.next
        print("None")  # Indicate end of queue
        
    def reverse(self):
        if self.is_empty():
            return "queue is empty"
        
        stack=[]
        while not self.is_empty():
            stack.append(self.dequeue())
        
        while stack:
            self.enqueue(stack.pop())
        
# Test the queue
a = Queue()     
a.enqueue(10)
print("After enqueue(10):")  # To differentiate outputs
a.display()

a.enqueue(20)
print("After enqueue(20):")
a.display()

a.enqueue(30)
print("After enqueue(30):")
a.display()
```

```python
class CircularQueue:
    def __init__(self, capacity):
        self.capacity = capacity
        self.queue = [None] * capacity
        self.front = -1
        self.rear = -1

    def is_empty(self):
        return self.front == -1

    def is_full(self):
        return (self.rear + 1) % self.capacity == self.front

    def enqueue(self, data):
        if self.is_full():
            print("Queue is full")
            return
        if self.is_empty():
            self.front = self.rear = 0
        else:
            self.rear = (self.rear + 1) % self.capacity
        self.queue[self.rear] = data

    def dequeue(self):
        if self.is_empty():
            print("Queue is empty")
            return
        print("Dequeued:", self.queue[self.front])
        if self.front == self.rear:
            self.front = self.rear = -1
        else:
            self.front = (self.front + 1) % self.capacity

    def get_front(self):
        if not self.is_empty():
            return self.queue[self.front] 
        else:
            "Queue is empty"

    def get_rear(self):
        if not self.is_empty():
            return self.queue[self.rear]
        else:
            "Queue is empty"

    def display(self):
        if self.is_empty():
            print("Queue is empty")
            return
        print("Circular Queue:", end=" ")
        i = self.front
        while True:
            print(self.queue[i], end=" ")
            if i == self.rear:
                break
            i = (i + 1) % self.capacity
        print()
        
cq = CircularQueue(5)
cq.enqueue(10)
cq.enqueue(20)
cq.enqueue(30)
cq.display() 
 
cq.dequeue()
cq.display()  

print("Front:", cq.get_front())  
print("Rear:", cq.get_rear())
```

CIRCULAR QUEUE

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularQueue:
    def __init__(self):
        self.front = None
        self.rear = None

    def is_empty(self):
        return self.front is None

    def enqueue(self, data):
        n = Node(data)
        if self.rear is None:
            self.front = self.rear = n
            self.rear.next = self.front  # Circular link
        else:
            self.rear.next = n
            self.rear = n
            self.rear.next = self.front  # Maintain circular link

    def dequeue(self):
        if self.is_empty():
            return "Queue is empty"
        
        temp = self.front.data  # Store front data to return

        if self.front == self.rear:  # Only one element in queue
            self.front = self.rear = None
        else:
            self.front = self.front.next
            self.rear.next = self.front  # Maintain circular link
        
        return temp

    def display(self):
        if self.is_empty():
            print("Queue is empty")
            return
        
        curr = self.front
        while curr:
            print(curr.data, end=" -> ")
            curr = curr.next
            if curr == self.front:  # Stop when full cycle is completed
                break
        print("(Back to front)")

# Test the Circular Queue
a = CircularQueue()
a.enqueue(10)
a.enqueue(20)
a.enqueue(30)

print("Queue after enqueue operations:")
a.display()

print("\nDequeued element:", a.dequeue())
print("Queue after one dequeue operation:")
a.display()
```

# DEQUE

```python
class node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class deque:
    def __init__(self):
        self.front = None
        self.rear = None

    def is_empty(self):
        return self.front is None

    def insert_front(self, data):
        n = node(data)
        if self.is_empty():
            self.front = self.rear = n
        else:
            n.next = self.front
            self.front.prev = n
            self.front = n

    def insert_rear(self, data):
        n = node(data)
        if self.is_empty():
            self.front = self.rear = n
        else:
            self.rear.next = n
            n.prev = self.rear
            self.rear = n

    def delete_front(self):
        if self.is_empty():
            return "queue is empty"
        
        temp = self.front.data
        if self.front == self.rear:
            self.front = self.rear = None
        else:
            self.front = self.front.next
            self.front.prev = None
        return temp

    def delete_rear(self):
        if self.is_empty():
            return "queue is empty"
        
        temp = self.rear.data
        if self.front == self.rear:
            self.front = self.rear = None
        else:
            self.rear = self.rear.prev
            self.rear.next = None
        return temp

    def display(self):
        if self.is_empty():
            print("queue is empty")
            return
        
        curr = self.front
        while curr:
            print(curr.data, end=" -> ")
            curr = curr.next
        print("None")  # Indicate end of deque

# ✅ Test the deque
dq = deque()
dq.insert_front(10)
dq.display()          # Output: 10 -> None

dq.insert_rear(20)
dq.display()          # Output: 10 -> 20 -> None

dq.insert_front(5)
dq.display()          # Output: 5 -> 10 -> 20 -> None

dq.insert_rear(30)
dq.display()          # Output: 5 -> 10 -> 20 -> 30 -> None
```

```python
class PriorityQueue:
    def __init__(self):
        self.items = []
        
    def is_empty(self):
        return len(self.items) == 0
    
    def push(self, data, priority_no):
        index = 0
        while index < len(self.items) and self.items[index][1] >= priority_no:
            index += 1
        self.items.insert(index, (data, priority_no))  # Insert in correct position
        
    def pop(self):
        if self.is_empty():
            raise IndexError("Priority queue is empty")
        
        return self.items.pop(0)[0]  # Pop highest priority element
        
    def size(self):
        return len(self.items)

# Testing the corrected PriorityQueue
p = PriorityQueue()
p.push("amit", 4)
p.push("ishant", 6)
p.push("ravi", 8)
p.push("manoj", 2)

while not p.is_empty():
    print(p.pop())  # Expected order: ravi, ishant, amit, manoj
```

```python
class Node:
    def __init__(self, data, priority):
        self.data = data
        self.priority = priority
        self.next = None

class PriorityQueueLL:
    def __init__(self):
        self.start = None  # Using 'start' instead of 'head'

    def is_empty(self):
        return self.start is None

    def push(self, data, priority):
        new_node = Node(data, priority)

        # If queue is empty or new node has higher priority than start
        if self.start is None or self.start.priority < priority:
            new_node.next = self.start
            self.start = new_node
        else:
            current = self.start
            while current.next is not None and current.next.priority >= priority:
                current = current.next
            new_node.next = current.next
            current.next = new_node

    def pop(self):
        if self.is_empty():
            raise IndexError("Priority queue is empty")
        
        popped_node = self.start
        self.start = self.start.next
        return popped_node.data

    def size(self):
        count = 0
        temp = self.start
        while temp:
            count += 1
            temp = temp.next
        return count

    def display(self):
        if self.is_empty():
            print("Priority Queue is empty")
        else:
            temp = self.start
            print("Priority Queue:")
            while temp:
                print(f"{temp.data} (Priority: {temp.priority})")
                temp = temp.next

# Example Testing
pq = PriorityQueueLL()
pq.push("amit", 4)
pq.push("ishant", 6)
pq.push("ravi", 8)
pq.push("manoj", 2)

pq.display()

print("\nPopping elements in priority order:")
while not pq.is_empty():
    print(pq.pop())
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):  
        for j in range(0, n - i - 1): 
            if arr[j] > arr[j + 1]:  
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  
    return arr

arr = [5, 4, 3, 2, 1]
print(bubble_sort(arr))  
```

```python
def insertion_sort(mylist):
    n=len(mylist)
    for i in range(1,n):
        temp=mylist[i]
        
        j=i-1
        
        while j>=0 and temp<mylist[j]:
            mylist[j+1]=mylist[j]
            j=j-1
            
        mylist[j+1]=temp
        
    return mylist
mylist=[5,1,6,3,7,9,4] 
insertion_sort(mylist)
        
```

```python
def selection_sort(arr):
    n=len(arr)
    for i in range(n):
        min_index=i
        for j in range(i+1,n):
            if arr[j]<arr[min_index]:
                min_index=j
        arr[i],arr[min_index]=arr[min_index],arr[i]
    
    return arr
arr=[5,4,37,90,0,-4,2,1]
selection_sort(arr)                       
```

```python
class Sorting:
    def __init__(self, arr):
        self.arr = arr  

    def bubble_sort(self):
        n = len(self.arr)
        for i in range(n):
            for j in range(n - i - 1):
                if self.arr[j] > self.arr[j + 1]:  
                    self.arr[j], self.arr[j + 1] = self.arr[j + 1], self.arr[j]
        return self.arr

    def insertion_sort(self):
        n = len(self.arr)
        for i in range(1, n):
            key = self.arr[i]
            j = i - 1
            while j >= 0 and self.arr[j] > key:
                self.arr[j + 1] = self.arr[j]  
                j -= 1
            self.arr[j + 1] = key  
        return self.arr

    def selection_sort(self):
        n = len(self.arr)
        for i in range(n):
            min_idx = i
            for j in range(i + 1, n):
                if self.arr[j] < self.arr[min_idx]:  
                    min_idx = j
            self.arr[i], self.arr[min_idx] = self.arr[min_idx], self.arr[i]  
        return self.arr


arr = [64, 25, 12, 22, 11]

sorter = Sorting(arr.copy())  

print("Bubble Sort:", sorter.bubble_sort())  
sorter.arr = arr.copy()  
print("Insertion Sort:", sorter.insertion_sort())  
sorter.arr = arr.copy()  
print("Selection Sort:", sorter.selection_sort())  
```

# HEAP 

```python
class MaxHeap:
    def __init__(self):
        self.heap = []

    # Insert an element and maintain the heap property
    def insert(self, val):
        self.heap.append(val)
        self.heapify_up(len(self.heap) - 1)

    # Move the inserted element up to its correct position
    def heapify_up(self, index):
        while index > 0:
            parent = (index - 1) // 2
            if self.heap[index] > self.heap[parent]:
                self.heap[index], self.heap[parent] = self.heap[parent], self.heap[index]
                index = parent
            else:
                break

    # Remove the root element and reheapify the heap
    def delete(self):
        if len(self.heap) == 0:
            print("Heap is empty")
            return
        # Replace root with last element
        self.heap[0] = self.heap[-1]
        self.heap.pop()
        self.heapify_down(0)

    # Move the root element down to maintain the heap property
    def heapify_down(self, index):
        size = len(self.heap)
        while index < size:
            left = 2 * index + 1
            right = 2 * index + 2
            largest = index

            if left < size and self.heap[left] > self.heap[largest]:
                largest = left
            if right < size and self.heap[right] > self.heap[largest]:
                largest = right

            if largest == index:
                break

            self.heap[index], self.heap[largest] = self.heap[largest], self.heap[index]
            index = largest

    # Display the heap
    def display(self):
        print("Heap:", self.heap)


# Test the MaxHeap
h = MaxHeap()
h.insert(10)
h.insert(20)
h.insert(5)
h.insert(30)
h.insert(15)
h.display()  # Shows the heap after insertions

h.delete()  # Removes the root (max element)
h.display()  # Shows the heap after deletion
```

```python
def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    # Compare left child with root
    if left < n and arr[left] > arr[largest]:
        largest = left

    # Compare right child with current largest
    if right < n and arr[right] > arr[largest]:
        largest = right

    # If largest is not root, swap and heapify the affected subtree
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)

    # Build max heap (rearrange array)
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    # Extract elements from heap one by one
    for i in range(n - 1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]  # Swap
        heapify(arr, i, 0)  # Heapify the reduced heap

# Example usage:
arr = [10, 3, 76, 34, 23, 32]
print("Original array:", arr)
heap_sort(arr)
print("Sorted array (Heap Sort):", arr)
```

```python
# Tree represented using dictionary
tree = {
    "A": ["B", "C", "D"],
    "B": ["E", "F"],
    "C": [],
    "D": ["G"],
    "E": [],
    "F": [],
    "G": []
}

# Pre-order Traversal (Root -> Children)
def pre_order(node):
    print(node, end=" ")
    for child in tree[node]:
        pre_order(child)

# Post-order Traversal (Children -> Root)
def post_order(node):
    for child in tree[node]:
        post_order(child)
    print(node, end=" ")

# Level-order Traversal (BFS)
def level_order(start):
    queue = [start]
    while queue:
        current = queue.pop(0)
        print(current, end=" ")
        for child in tree[current]:
            queue.append(child)

# Driver Code
print("Preorder Traversal:")
pre_order("A")     # Output: A B E F C D G
print("\n")

print("Postorder Traversal:")
post_order("A")    # Output: E F B C G D A
print("\n")

print("Level Order Traversal:")
level_order("A")   # Output: A B C D E F G
```

```python
class Tree:
    def __init__(self,data):
        self.data=data
        self.children=[]

    def pre_order(self):
        print(self.data,end=" ")
        for child in self.children:
            child.pre_order()
            
    def post_order(self):
        for child in self.children:
            child.post_order()
        print(self.data,end=" ")
            
    def level_order(self):
        queue=[self]
        
        while queue:
            curr=queue.pop(0)
            print(curr.data,end=" ")
            
            for child in curr.children:
                queue.append(child)
                

root=Tree("A")
b=Tree("B")
c=Tree("C")
d=Tree("D")
e=Tree("E")
f=Tree("F")
g=Tree("G")

root.children.extend([b,c,d])
b.children.extend([e,f])
d.children.append(g)



print("Preorder Traversal:")
root.pre_order()        # Output: A B E F C D G
print("\n")

print("Postorder Traversal:")
root.post_order()       # Output: E F B C G D A
print("\n")
   
print("Level Order Traversal:")
root.level_order()
```

```python
class Node:
    def __init__(self, item):
        self.item = item
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def insert(self, data):
        self.root = self.rinsert(self.root, data)

    def rinsert(self, root, data):
        if root is None:
            return Node(data)
        if data < root.item:
            root.left = self.rinsert(root.left, data)
        elif data > root.item:
            root.right = self.rinsert(root.right, data)
        return root

    def search(self, data):
        return self.rsearch(self.root, data)

    def rsearch(self, root, data):
        if root is None or root.item == data:
            return root
        if data < root.item:
            return self.rsearch(root.left, data)
        else:
            return self.rsearch(root.right, data)

    def in_order_traversal(self):
        result = []
        self.rinorder(self.root, result)
        return result

    def rinorder(self, root, result):
        if root:
            self.rinorder(root.left, result)
            result.append(root.item)
            self.rinorder(root.right, result)

    def pre_order_traversal(self):
        result = []
        self.rpreorder(self.root, result)
        return result

    def rpreorder(self, root, result):
        if root:
            result.append(root.item)
            self.rpreorder(root.left, result)
            self.rpreorder(root.right, result)

    def post_order_traversal(self):
        result = []
        self.rpostorder(self.root, result)
        return result

    def rpostorder(self, root, result):
        if root:
            self.rpostorder(root.left, result)
            self.rpostorder(root.right, result)
            result.append(root.item)

    def min_val(self, temp):
        curr = temp
        while curr.left is not None:
            curr = curr.left
        return curr.item

    def max_val(self, temp):
        curr = temp
        while curr.right is not None:
            curr = curr.right
        return curr.item

    def delete(self, data):
        self.root = self.rdelete(self.root, data)

    def rdelete(self, root, data):
        if root is None:
            return root
        if data < root.item:
            root.left = self.rdelete(root.left, data)
        elif data > root.item:
            root.right = self.rdelete(root.right, data)
        else:
            # Node with only one child or no child
            if root.left is None:
                return root.right
            elif root.right is None:
                return root.left
            # Node with two children
            root.item = self.min_val(root.right)
            root.right = self.rdelete(root.right, root.item)
        return root

    def size(self):
        return len(self.in_order_traversal())

# ---------------------------
# TESTING WITH PRINTS
# ---------------------------
a = BST()

print("Initial Tree (should be empty):")
print("In-Order Traversal:", a.in_order_traversal())
print("Pre-Order Traversal:", a.pre_order_traversal())
print("Post-Order Traversal:", a.post_order_traversal())
print()

# Inserting basic nodes
a.insert(70)
a.insert(10)
a.insert(90)

print("Tree after inserting 70, 10, 90:")
print("In-Order Traversal:", a.in_order_traversal())
print("Pre-Order Traversal:", a.pre_order_traversal())
print("Post-Order Traversal:", a.post_order_traversal())
print()

# Inserting more nodes
a.insert(25)
a.insert(100)
a.insert(60)
a.insert(40)
a.insert(45)

print("Tree after inserting 25, 100, 60, 40, 45:")
print("In-Order Traversal:", a.in_order_traversal())
print("Pre-Order Traversal:", a.pre_order_traversal())
print("Post-Order Traversal:", a.post_order_traversal())
print()

print("Minimum value in the BST:", a.min_val(a.root))
print("Maximum value in the BST:", a.max_val(a.root))
print()

# Deleting a node (leaf node 45)
print("Deleting node 45...")
a.delete(45)

print("Tree after deleting 45:")
print("In-Order Traversal:", a.in_order_traversal())
print("Tree Size:", a.size())
```
